```python
import pandas as pd
import requests
from selenium.webdriver.chrome.options import Options  
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from google import google #Google api
try:
    from bs4 import BeautifulSoup
except ImportError:
    from BeautifulSoup import BeautifulSoup
#Install all the required modules
```

```python
num_page = 3
to_search ="site:linkedin.com/in/ AND Clinical Trial Manager AND Pharmaceuticals AND UK"
search_results = google.search(to_search, num_page)
#to_search, put job title after first AND and place after Second AND
```

```python
links = []
for i in search_results:
    links.append(i.link)
```

```python
data_dict={}
name=[]
location=[]
companay=[]
from linkedin_scraper import Person, actions
driver = webdriver.Chrome()
email = "" #Enter your linkdin email
password = ""#Enter your linkdin password
actions.login(driver, email, password) # if email and password isnt given, it'll prompt in terminal
for i in links:
    driver.get(i)
    soup_linkdin = BeautifulSoup(driver.page_source, 'html.parser')
    name.append(soup_linkdin.find('li',{"class":"inline t-24 t-black t-normal break-words"}).text.strip())
    location.append(soup_linkdin.find('li',{"class":"t-16 t-black t-normal inline-block"}).text.strip())
    companay.append(soup_linkdin.find('span',{"class":"text-align-left ml2 t-14 t-black t-bold full-width lt-line-clamp lt-line-clamp--multi-line ember-view"}).text.strip())
data_dict={
    "Name":name,
    "Location":location,
    "Companay":companay,
}
pd.DataFrame(data_dict)    
    
```
